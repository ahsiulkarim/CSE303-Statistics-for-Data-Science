```python
# 1 lode data

import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid")
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scipy import stats

df = pd.read_csv('/content/drive/MyDrive/CSE303  Practice/CSE303  Practice/Fall 2025/CSE303_Lab4:Basic Ploting/Lab4_task/Walmart_sales.csv')
df
```

```python
# 2.Count Plot of Holiday vs. Non-Holiday Weeks:

plt.figure(figsize=(10,8), dpi=77)
sns.countplot(x="Holiday_Flag" ,data=df)
plt.title("Count Plot of Holiday vs. Non-Holiday Weeks")
plt.xlabel("Holiday Flag")
plt.ylabel("Count")
plt.show()
```

```python
#3.Bar Plot of Total Sales by Store
plt.figure(figsize=(10,8), dpi=77)
sns.barplot(x="Store", y="Weekly_Sales", data=df)
plt.title("Bar Plot of Total Sales by Store")
plt.xlabel("Store")
plt.ylabel("Total Sales")
plt.show()
```

```python
# 4.Pie Chart of Holiday vs. Non-Holiday Sales Proportion:
holiday_sales = df[df['Holiday_Flag'] == 1]['Weekly_Sales'].sum()
non_holiday_sales = df[df['Holiday_Flag'] == 0]['Weekly_Sales'].sum()
proportions = [holiday_sales, non_holiday_sales]
labels = ['Holiday', 'Non-Holiday']
plt.figure(figsize=(6, 6))
plt.pie(proportions, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart of Holiday vs. Non-Holiday Sales Proportion")
plt.show()
```

```python
# 5. Count Plot of Sales by Month:
df['Month'] = pd.to_datetime(df['Date'], format='%d-%m-%Y').dt.month
plt.figure(figsize=(10,6))
plt.title("Count Plot of Sales by Month")
plt.xlabel("Month")
plt.ylabel("Count")
sns.countplot(x="Month",data = df)
plt.show()
```

```python
# 6. Bar Plot of Total Sales by Month:
plt.figure(figsize=(10,6))
sns.barplot(x='Month',y='Weekly_Sales',data=df)
plt.title("Bar Plot of Total Sales by Month")
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Pie Chart of Sales Proportion by Month:

# Convert Date column to datetime and extract month
#df['Month'] = pd.to_datetime(df['Date'], format='%d-%m-%Y').dt.month

# Group by Month and sum the sales
monthly_sales = df.groupby('Month')['Weekly_Sales'].sum()

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie( monthly_sales,labels=monthly_sales.index,autopct='%1.1f%%',startangle=90)
plt.title('Pie Chart of Sales Proportion by Month')
plt.show()
```
