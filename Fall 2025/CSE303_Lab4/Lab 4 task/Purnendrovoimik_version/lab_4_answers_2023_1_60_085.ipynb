```python
# 1. Load Data
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid")
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scipy import stats

task_data = pd.read_csv('/content/drive/MyDrive/EWU_Study/CSE303/DMIM/Labs/Lab 4/Task/Walmart_sales.csv')

task_data
```

```python
# 2. Count Plot of Holiday vs. Non-Holiday Weeks:

plt.figure(figsize=(10, 6), dpi = 77)
sns.countplot(x='Holiday_Flag', data=task_data)
plt.title('Count Plot of Holiday vs. Non-Holiday Weeks')
plt.xlabel('Holiday Flag')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Bar Plot of Total Sales by Store:
plt.figure(figsize=(10, 8), dpi = 77)
sns.barplot(x='Store', y='Weekly_Sales', data=task_data)
plt.title('Bar Plot of Total Sales by Store')
plt.xlabel('Store')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 4. Pie Chart of Holiday vs. Non-Holiday Sales Proportion:
holiday_sales = task_data[task_data['Holiday_Flag'] == 1]['Weekly_Sales'].sum()
non_holiday_sales = task_data[task_data['Holiday_Flag'] == 0]['Weekly_Sales'].sum()
proportions = [holiday_sales, non_holiday_sales]
labels = ['Holiday', 'Non-Holiday']
plt.figure(figsize=(6, 6))
plt.pie(proportions, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Holiday vs. Non-Holiday Sales Proportion')
plt.show()
```

```python
# 5. Count Plot of Sales by Month:
task_data['Month'] = pd.to_datetime(task_data['Date'], format='%d-%m-%Y').dt.month
plt.figure(figsize=(10, 6))
plt.title('Count Plot of Sales by Month')
plt.xlabel('Month')
plt.ylabel('Count')
sns.countplot(x='Month', data=task_data)
plt.show()
```

```python
# 6. Bar Plot of Total Sales by Month:
plt.figure(figsize=(10, 6))
plt.title('Bar Plot of Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
sns.barplot(x='Month', y='Weekly_Sales', data=task_data)
plt.show()
```

```python
# 7. Pie Chart of Sales Proportion by Month:
monthly_sales = task_data.groupby('Month')['Weekly_Sales'].sum()
proportions = monthly_sales / monthly_sales.sum()
labels = monthly_sales.index
plt.figure(figsize=(6, 6))
plt.pie(proportions, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Sales by Month')
plt.show()
```
